## Introducción

Los **filtros pasivos RLC** permiten modificar la amplitud y fase de una señal en función de la frecuencia, seleccionando o atenuando ciertas bandas del espectro.  
Según cómo se conecten los elementos resistivos, inductivos y capacitivos, el circuito puede comportarse como **pasa–bajos**, **pasa–altos** o **pasa–banda**.

En el análisis en el dominio de Laplace, la relación entre las tensiones de entrada y salida se expresa como:

$$
T(s) = \frac{V_2(s)}{V_1(s)}
$$

Para el circuito propuesto, dicha función adopta la forma **bi–cuadrática**:

$$
T(s) = k \cdot \frac{s^2 + \frac{\omega_{0z}}{Q_z}s + \omega_{0z}^2}{s^2 + \frac{\omega_{0p}}{Q_p}s + \omega_{0p}^2}
$$

donde:
- $k$ es la **ganancia** de continua,  
- $\omega_{0p}$ y $\omega_{0z}$ son las **frecuencias naturales** de polos y ceros,  
- y $Q_p$, $Q_z$ representan los **factores de calidad**, que determinan la selectividad y el ancho de banda del filtro.

El **ancho de banda** efectivo se relaciona directamente con el factor de calidad:

$$
BW = \frac{\omega_0}{Q}
$$

A lo largo del trabajo se buscará caracterizar experimental y teóricamente las tres funciones propuestas:

$$
T_1(s), \quad T_2(s), \quad T_3(s)
$$

analizando sus diagramas de polos y ceros, así como sus respuestas de magnitud y fase, con el objetivo de comprender cómo los parámetros del circuito afectan la forma final del filtro.

### Análisis de $T_1(s)$

$$
T_1(s) = \frac{s^2 + 9}{s^2 + s\sqrt{2} + 1}
$$

**Polos:**

$$
s^2 + s\sqrt{2} + 1 = 0 \Rightarrow s = \frac{-\sqrt{2} \pm \sqrt{2-4}}{2} = -\frac{\sqrt{2}}{2} \pm j\frac{\sqrt{2}}{2}
$$

$\Rightarrow$ Polos complejos conjugados, con $\omega_{0p} = 1$ y $Q_p = \frac{1}{\sqrt{2}} \approx 0,707$

**Ceros:**

$$
s^2 + 9 = 0 \Rightarrow s = \pm 3j
$$

$\Rightarrow$ Frecuencia de cero $\omega_{0z} = 3$ (sin término lineal, $Q_z \to \infty$)


$T_1(s)$ tiene ceros puros imaginarios más alejados que los polos, actuando como un filtro rechaza banda (notch) atenuado en bajas frecuencias, según la ganancia $k$.


### Análisis de $T_2(s)$

$$
T_2(s) = \frac{s^2 + \frac{1}{9}}{s^2 + s\frac{1}{5} + 1}
$$

**Polos:**

$$
s^2 + \frac{1}{5}s + 1 = 0
$$

$\Rightarrow$ Polos complejos conjugados con $\omega_{0p} = 1$ y $Q_p = 2,5$

**Ceros:**

$$
s^2 + \frac{1}{9} = 0 \Rightarrow s = \pm \frac{j}{3}
$$

$\Rightarrow$ Frecuencia de cero $\omega_{0z} = \frac{1}{3}$ (sin término lineal, $Q_z \to \infty$)


$T_2(s)$ tiene ceros a baja frecuencia y polos más próximos a la unidad, por lo que se comporta como un filtro pasa altos, atenuando las bajas frecuencias.


### Análisis de $T_3(s)$

$$
T_3(s) = \frac{s^2 + s\frac{1}{5} + 1}{s^2 + s\sqrt{2} + 1}
$$

**Numerador:**

$$
s^2 + 0,2s + 1 \Rightarrow \omega_{0z} = 1, \quad Q_z = \frac{1}{2 \times 0.2} = 2,5
$$

**Denominador:**

$$
s^2 + s\sqrt{2} + 1 \Rightarrow \omega_{0p} = 1, \quad Q_p = \frac{1}{\sqrt{2}} \approx 0,707
$$


$T_3(s)$ comparte los polos de $T_1(s)$ y ambos son un filtro rechaza banda, pero $T_2$ tiene ceros con un factor de calidad mayor ($Q_z = 2,5$), por lo que modifica la resonancia del denominador, suavizando el pico en la respuesta en frecuencia.

4. Dibujamos el diagrama de polos y ceros de T(s)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.patches as patches
```

### Diagrama de polos y ceros para $T_1(s)$:

```python
#T1(s) = (s^2 + 9) / (s^2 + sqrt(2)s + 1)
num1 = [1, 0, 9]
den1 = [1, np.sqrt(2), 1]
z1, p1, k1 = signal.tf2zpk(num1, den1)

plt.figure(figsize=(6,6))
plt.plot(np.real(p1), np.imag(p1), 'x', markersize=10, label='T1 Polos', color='orchid', markeredgewidth=2.5)
if len(z1) > 0:
    plt.plot(np.real(z1), np.imag(z1), 'o', markersize=10, fillstyle='none', label='T1 Ceros', color='lightblue', markeredgewidth=2.5)

# Círculo unitario
unit_circle = patches.Circle((0, 0), radius=1, fill=False,
                             color='gray', ls='dotted', lw=2)

axes_hdl = plt.gca()
axes_hdl.add_patch(unit_circle)

plt.axis([-1.1, 1.1, -3.5, 3.5])
plt.axis('equal')
plt.title('T1 - Diagrama de Polos y Ceros (plano s)')
plt.xlabel(r'$\Re(s)$')
plt.ylabel(r'$\Im(s)$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

- En el plano $s$, los polos se ubican cerca del eje imaginario, con una pequeña parte real negativa, por lo que el sistema es estable.  
- Los ceros están más alejados del origen (en $s = \pm j3$), lo que sugiere que afectan principalmente las frecuencias más altas.  
- Como los polos están cerca de $s = \pm j1$ y los ceros lejos, la respuesta se atenúa en torno a la frecuencia de los polos pero mantiene ganancia alta en bajas y altas frecuencias.  
- Por esto, el comportamiento observado es el de un filtro rechaza-banda (notch).

- En consecuencia, el comportamiento observado corresponde a un filtro rechaza banda suave (notch leve), que reduce la ganancia en la zona central del espectro pero mantiene la estabilidad y una transición gradual entre bandas.

5. Calculamos y graficamos la respuesta de módulo y fase, indicando la respuesta asintótica donde corresponda. 

```python
#-----------FUncion hecha para calcular la repsuesta asintótica-----------#
def bode_asintotica(w, quiebres, nivel_baja_db):
    y = np.full_like(w, nivel_baja_db, dtype=float)
    qs = sorted(quiebres, key=lambda x: x[0])
    slope = 0.0
    last_w = w[0]
    last_y = nivel_baja_db
    qi = 0
    for i, wi in enumerate(w):
        while qi < len(qs) and wi >= qs[qi][0]:
            w_break, delta = qs[qi]
            if w_break > last_w:
                last_y += slope * np.log10(w_break/last_w)
                last_w = w_break
            slope += delta
            qi += 1
        y[i] = last_y + slope * np.log10(wi/last_w) if wi >= last_w else last_y
    return y
```

### Respuesta de Módulo y Fase para $T_1(s)$:

```python
w1, h1 = signal.freqs(num1, den1, worN=np.logspace(-1, 2, 1000))
phase1 = np.unwrap(np.angle(h1))
gd1 = -np.diff(phase1) / np.diff(w1)

plt.figure(figsize=(12, 8))
plt.subplot(2,1,1)
plt.semilogx(w1, 20*np.log10(abs(h1)), label='T1(s)', color='lightblue')
asint1 = bode_asintotica(w1, [(1.0, -40.0), (3.0, +40.0)], 20*np.log10(9.0))
plt.semilogx(w1, asint1, '--', color='orchid', label='Asintótica')
plt.title('Magnitud - T1(s)')
plt.xlabel('ω [rad/s]')
plt.ylabel('|H(jω)| [dB]')
plt.grid(True, which='both', ls=':')
plt.legend()

plt.subplot(2,1,2)
plt.semilogx(w1, np.degrees(phase1), label='T1(s)', color="orchid")
plt.title('Fase - T1(s)')
plt.xlabel('ω [rad/s]')
plt.ylabel('Fase [°]')
plt.grid(True, which='both', ls=':')
plt.legend()
plt.tight_layout()
plt.show()
```

$$T_1(s) = \dfrac{s^2 + 9}{s^2 + s\sqrt{2} + 1}$$

- A bajas y altas frecuencias, la ganancia se mantiene relativamente constante y alta (cercana a 0 dB o positiva)
- La pendiente asintótica muestra un descenso de $-40$ dB/década en la frecuencia de quiebre y una recuperación de $+40$ dB/década en la siguiente, lo cual indica la presencia de una banda atenuada central.  
- La fase atraviesa una variación significativa, típica de un rechaza-banda.  
- Por lo tanto, la respuesta en frecuencia confirma que $T_1(s)$ actúa como un filtro rechaza-banda (notch), con una atenuación marcada en torno a su frecuencia natural y ganancia alta fuera de esa región.

# BONUS

Conectar dos funciones de transferencia en cascada significa que la salida de la primera se convierte en la entrada de la segunda., es decir, se multiplican: 
$$T_{\text{cascada}}(s) = T_1(s) \cdot T_2(s)$$

```python
num_total = np.polymul(num1, num2)
den_total = np.polymul(den1, den2)

# Respuesta en frecuencia
w = np.logspace(-1, 2, 1000)
w_eval, h_total = signal.freqs(num_total, den_total, worN=w)
phase_total = np.unwrap(np.angle(h_total))

# Quiebres: polos en ω=1, ceros en ω=1/3 y ω=3
quiebres = [(1/3, +40.0), (1.0, -40.0), (3.0, +40.0)]
nivel_base_db = 20*np.log10(9 * (1/9))  # Ganancia en baja frecuencia
asint_total = bode_asintotica(w, quiebres, nivel_base_db)

#Gráfico de módulo y fase
plt.figure(figsize=(12, 8))

# Módulo
plt.subplot(2,1,1)
plt.semilogx(w_eval, 20*np.log10(abs(h_total)), label='T1(s)·T2(s)', color='lightblue')
plt.semilogx(w, asint_total, '--', color='orchid', label='Asintótica')
plt.title('Magnitud - T1(s)·T2(s)')
plt.xlabel('ω [rad/s]')
plt.ylabel('|H(jω)| [dB]')
plt.grid(True, which='both', ls=':')
plt.legend()

# Fase
plt.subplot(2,1,2)
plt.semilogx(w_eval, np.degrees(phase_total), label='T1(s)·T2(s)', color='orchid')
plt.title('Fase - T1(s)·T2(s)')
plt.xlabel('ω [rad/s]')
plt.ylabel('Fase [°]')
plt.grid(True, which='both', ls=':')
plt.legend()

plt.tight_layout()
plt.show()
```

- La cascada de dos filtros rechaza banda genera una respuesta compuesta que atenúa en dos regiones distintas del espectro: una en torno a $\omega = \frac{1}{3}$ y otra en $\omega = 3$.
- La respuesta total tiene una forma más compleja, con dos quiebres descendentes y una recuperación en alta frecuencia.
- La ganancia en baja frecuencia es el producto de las ganancias individuales:
$|H(j0)| = \left| \frac{9}{1} \cdot \frac{1/9}{1} \right| = 1 \Rightarrow 0 \text{ dB}$
- El sistema está normalizado en baja frecuencia.
- La conexión en cascada multiplica las funciones de transferencia, lo que permite construir filtros más selectivos o con múltiples bandas de rechazo.
