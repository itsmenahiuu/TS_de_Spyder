## Introducción

En sistemas discretos LTI, la relación entre la entrada $x(n)$ y la salida $y(n)$ puede escribirse como una ecuación en diferencias.  
Al aplicar la **Transformada Z**, se obtiene una relación algebraica:

$$
T(z) = \frac{Y(z)}{X(z)}
$$

donde $T(z)$ es la función de transferencia del sistema. Describe cómo cada componente frecuencial de la entrada se modifica en amplitud y fase.

En los filtros FIR, $T(z)$ es un polinomio en potencias negativas de $z$ (solo términos de entrada); por eso son no recursivos y estables, y suelen presentar fase casi lineal.

Para estudiar el comportamiento espectral se evalúa $T(z)$ en el círculo unidad:

$$
z = e^{j\omega}
$$

y así se obtiene la respuesta en frecuencia:

$$
H(e^{j\omega}) = T(z)\bigg|_{z=e^{j\omega}}
$$

- **Módulo:** $|H(e^{j\omega})|$ indica la ganancia/atenuación por frecuencia.  
- **Fase:** $\angle H(e^{j\omega})$ indica el desplazamiento de fase.

En general, los **promedios móviles** atenúan altas frecuencias (pasa-bajos) y las **diferencias** realzan cambios rápidos (pasa-altos).

**Análisis de resultados – Punto 1**

- Los sistemas (a) y (b) son filtros de *media móvil*.  
  Promedian varias muestras consecutivas de la señal de entrada, reduciendo las variaciones rápidas y el ruido.  
  Por eso actúan como *filtros pasa-bajos*, que dejan pasar las bajas frecuencias y atenúan las altas.

- Los sistemas (c) y (d) son *filtros diferencia*.  
  Calculan la diferencia entre una muestra y otra separada por uno o dos períodos.

**Conclusión:**

Los cuatro sistemas son **FIR**, es decir, tienen una duración finita de respuesta al impulso y no utilizan realimentación.  
Por esta razón, son *incondicionalmente estables* y de fácil implementación.  
Mientras que los de media móvil suavizan la señal, los de diferencia permiten detectar variaciones bruscas o bordes en la señal de entrada.

### Punto 2 – Respuesta en frecuencia $$H(e^{j\omega})$$

Para estudiar el comportamiento espectral de cada sistema, se evalúa la función de transferencia sobre el **círculo unidad** del plano Z, reemplazando:

$$
z = e^{j\omega}
$$

De esta forma, se obtiene la **respuesta en frecuencia** del sistema:

$$
H(e^{j\omega}) = T(z)\big|_{z=e^{j\omega}}
$$

**Análisis de resultados – Punto 2**

- **Sistemas (a) y (b) – Media móvil (pasa-bajos)**
  
  El módulo presenta un máximo en
  $$
  \omega = 0,
  $$
  y disminuye a medida que crece la frecuencia. Esto indica que se *preservan las bajas frecuencias* y se *atenuan las altas*.  
  Además aparecen ceros periódicos sobre el círculo unidad, generando frecuencias donde
  $$
  |H(e^{j\omega})| = 0.
  $$

- **Sistemas (c) y (d) – Diferencia (pasa-altos)**
  
  El módulo es nulo en
  $$
  \omega = 0
  $$
  y crece hasta cercanías de
  $$
  \omega = \pi,
  $$
  lo que refleja que *se cancelan componentes lentas y/o constantes* y *se realzan variaciones rápidas*.

- **Fase en todos los sistemas**
  
  La fase es aproximadamente lineal con la frecuencia, por lo que el sistema introduce *retardo casi constante*.

**Conclusión:**

- (a) y (b): filtros de **media móvil** ⇒ comportamiento **pasa-bajos**.  
- (c) y (d): filtros de **diferencia** ⇒ comportamiento **pasa-altos**.  
- En todos los casos, la fase **lineal** implica un desplazamiento temporal consistente sin deformar la forma de onda.

### Punto 3 – Simulación y validación numérica (NumPy)


Para validar los resultados del Punto 2 se calcula numéricamente la respuesta en frecuencia evaluando:

$$
H(e^{j\omega}) = \sum_{k=0}^{M} b_{k}\, e^{-j\omega k}
$$

donde los coeficientes $b_{k}$ corresponden a cada filtro FIR.

Se grafican el **módulo** y la **fase** de los sistemas (a), (b), (c) y (d), y se comparan con las expresiones analíticas:

$$
\begin{aligned}
|H_a(e^{j\omega})| &= \big|\,2\cos\tfrac{3\omega}{2} + 2\cos\tfrac{\omega}{2}\,\big|,\\[4pt]
|H_b(e^{j\omega})| &= \big|\,1 + 2\cos(\omega) + 2\cos(2\omega)\,\big|,\\[4pt]
|H_c(e^{j\omega})| &= 2\big|\sin\tfrac{\omega}{2}\big|,\\[4pt]
|H_d(e^{j\omega})| &= 2|\sin(\omega)|.
\end{aligned}
$$


----

### Codigo


```python
import numpy as np
import matplotlib.pyplot as plt

#malla de frecuencia
w = np.linspace(-np.pi, np.pi, 2000)
z = np.exp(-1j * w)

#coeficientes de mi FIR
b_a = np.array([1,1,1,1])       # a) y(n)=x(n−3)+x(n−2)+x(n−1)+x(n) 
b_b = np.array([1,1,1,1,1])     # b) x(n−4)+x(n−3)+x(n−2)+x(n−1)+x(n) 
b_c = np.array([1,-1])          # c) y(n)=x(n)−x(n−1) 
b_d = np.array([1,0,-1])        # d) y(n)=x(n)−x(n−2) 


# H(e^{jω})
H_a = np.polyval(b_a[::-1], z) #np.polyval evalua el polinomio en z = e^{-jω}
H_b = np.polyval(b_b[::-1], z)
H_c = np.polyval(b_c[::-1], z)
H_d = np.polyval(b_d[::-1], z)

#fases
fase_a = np.unwrap(np.angle(H_a))
fase_b = np.unwrap(np.angle(H_b))
fase_c = np.unwrap(np.angle(H_c))
fase_d = np.unwrap(np.angle(H_d))

# =======================================================
# a)

plt.figure(1)
plt.title("a) Módulo")
plt.plot(w, np.abs(H_a), color = "mediumorchid")
plt.ylabel("|H|")
plt.grid(True)

plt.subplot(2,1,2)
plt.title("a) Fase")
plt.plot(w, fase_a, color = "mediumorchid")
plt.xlabel("ω [rad]")
plt.ylabel("Fase [rad]")
plt.grid(True)

plt.tight_layout()
plt.show()

# =======================================================
# b) 

plt.figure(2)

plt.subplot(2,1,1)
plt.title("b) Módulo")
plt.plot(w, np.abs(H_b), color = "mediumorchid")
plt.ylabel("|H|")
plt.grid(True)

plt.subplot(2,1,2)
plt.title("b) Fase")
plt.plot(w, fase_b, color = "mediumorchid")
plt.xlabel("ω [rad]")
plt.ylabel("Fase [rad]")
plt.grid(True)

plt.tight_layout()
plt.show()

# =======================================================
# c)

plt.figure(3)
plt.subplot(2,1,1)
plt.title("c) Módulo")
plt.plot(w, np.abs(H_c), color = "mediumorchid")
plt.ylabel("|H|")
plt.grid(True)

plt.subplot(2,1,2)
plt.title("c) Fase")
plt.plot(w, fase_c, color = "mediumorchid")
plt.xlabel("ω [rad]")
plt.ylabel("Fase [rad]")
plt.grid(True)

plt.tight_layout()
plt.show()

# =======================================================
# d)

plt.figure(4)
plt.subplot(2,1,1)
plt.title("d) Módulo")
plt.plot(w, np.abs(H_d), color = "mediumorchid")
plt.ylabel("|H|")
plt.grid(True)

plt.subplot(2,1,2)
plt.title("d) Fase")
plt.plot(w, fase_d, color = "mediumorchid")
plt.xlabel("ω [rad]")
plt.ylabel("Fase [rad]")
plt.grid(True)

plt.tight_layout()
plt.show()
```

---
**Análisis de resultados – Punto 3**

- En los filtros **(a)** y **(b)** (media móvil), el módulo presenta un valor máximo en  

  $$
  \omega = 0
  $$

  y decae para frecuencias más altas, confirmando su comportamiento *pasa-bajos*.  
  Además, se observan *ceros periódicos* en el espectro, que anulan la respuesta en determinadas frecuencias.  
  Esto coincide con los resultados teóricos del Punto 2.

- En los filtros **(c)** y **(d)** (diferencia), el módulo es nulo en  

  $$
  \omega = 0
  $$

  y aumenta hasta alcanzar su máximo en  

  $$
  \omega = \pi
  $$

  mostrando un comportamiento *pasa-altos*.  
  Este tipo de respuesta amplifica las transiciones rápidas o los cambios abruptos de la señal.

- En todos los casos, la *fase* observada es lineal con la frecuencia, lo cual implica un *retardo constante*.  
  Esto significa que los filtros no distorsionan la forma de onda de la señal, sólo la desplazan temporalmente.
