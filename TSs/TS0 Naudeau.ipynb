## CÃ³digo

```python
import numpy as np  
import matplotlib.pyplot as plt

def mi_funcion_sen( vmax, dc, ff, ph, nn, fs): 
#defino (def) mi funcion para pasarle todos los datos

#vmax:amplitud max de la senoidal [Volts]
#dc:valor medio [Volts]
#ff:frecuencia [Hz]
#ph:fase en [rad]
#nn:cantidad de muestras
#fs:frecuencia de muestreo [Hz]

    Ts = 1/fs #perÃ­odo de muestre (osea cuanto tiempo hay entre muestra y muestra)
    tt = np.linspace(0, (nn-1)*Ts, nn) #tt seria el vector de tiempos donde se evalÃºa la seÃ±al
#np.linspace(start, stop, num, endpoint=True, retstep=False, dtype=None, axis=0)

# - start: valor inicial
# - stop: valor final
# - num: cantidad de puntos (default = 50)
# - endpoint: True -> incluye el valor final; False -> no lo incluye
# - retstep: True -> devuelve tambien el paso entre puntos
# - dtype: tipo de dato (float, int, etc.)
# - axis: eje en el que se coloca el resultado en arrays multidimensionales
 
    xx=vmax*np.sin(2*np.pi*ff*tt+ph)+dc #np.pi es pi y np.sin es para hacer el seno

    return tt,xx #cierro mi funcion "def", es como las funciones de C pero no tiene llaves, es asi nomas
#se tabula todo lo que esta dentro de mi funcion 

N=1000
fs=1000
tt, yy = mi_funcion_sen(1, 0, 1, 0, N, fs)
plt.figure(1)
plt.plot(tt, yy, color='orchid') 
plt.title('SeÃ±al Senoidal')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [Volts]')
plt.grid(True)
plt.show()
```

## Conclusiones

Los grÃ¡ficos muestran cÃ³mo la frecuencia de la seÃ±al afecta su representaciÃ³n muestreada.
- Cuando la frecuencia de mi seÃ±al es menor a la mitad de la frecuencia de muestreo (cantidad de veces por segundo que se toma una muestra de una seÃ±al continua para convertirla en digital) la onda senoidal se ve correctamente. Esto demuestra que el muestreo cumple con el teorema de Nyquist, por lo que la seÃ±al puede ser reconstruida sin perder informaciÃ³n.
                          ***frecuencia de nyquist*** --> $\frac{f_s}{2}$
- Cuando utilizo la frecuencia de Nyquist, teÃ³ricamente deberÃ­a poder armar una senoidal (con 2 muestras --> $\frac{f_s}{f_f}$), pero realmente no se puede distinguir, lo que hace el cÃ³digo es poner mis dos muestras en donde la senoidal vale cero e intentar gracficar algo, dando como resultado un grafico nada parecido a una senoidal con valores del eje "y" muy cercanos a cero.
Si yo quisiese ver los picos de la amplitud mÃ¡xima, puedo defasar mi onda al sumar $\frac{\pi}{2}$
- Cuando mi frecuencia esta por encima de nyquist se produce un solapamiento o *"aliasing"* donde mi muestreo ya no me puede armar mi senoidal y las replicas espectrales se solapan. Cuando estas se solapan nos queda una seÃ±al "invertida" con una frecuencia aparente donde:
$$f_{aparente}=|f_s - f_{seÃ±al}|$$
